```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt

# Create a 2nd degree polynomial dataset
X = np.linspace(-10, 10, 1000)
Y = X**2

# Reshape for training
X = X.reshape(-1, 1)
Y = Y.reshape(-1, 1)

# Define a model
model = Sequential()
model.add(Dense(3, input_dim=1, activation='relu'))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')

# Train the model
model.fit(X, Y, epochs=500, verbose=0)

# Predict the output
Y_pred = model.predict(X)

# Plot the original function and the model's predictions
plt.figure(figsize=(8,6))
plt.plot(X, Y, label='Original Function')
plt.plot(X, Y_pred, label='Neural Network Output')

# Show the activations of each neuron in the hidden layer for the input range
hidden_layer_weights = model.layers[0].get_weights()[0]
hidden_layer_biases = model.layers[0].get_weights()[1]

for i in range(hidden_layer_weights.shape[1]):
    neuron_output = np.maximum(0, X.dot(hidden_layer_weights[:,i]) + hidden_layer_biases[i])
    plt.plot(X, neuron_output, label=f'Neuron {i+1} Output')

#plt.xlim[0,100]
plt.legend()
plt.show()
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LeakyReLU
from tensorflow.keras.initializers import HeNormal
import matplotlib.pyplot as plt

# Create a 2nd degree polynomial dataset
X = np.linspace(0, 10, 1000)
Y = X**2

# Reshape for training
X = X.reshape(-1, 1)
Y = Y.reshape(-1, 1)

# Define a model
model = Sequential()

# Use He Normal initializer
initializer = HeNormal()

# Use LeakyReLU activation
model.add(Dense(3, input_dim=1, kernel_initializer=initializer))
model.add(LeakyReLU())
model.add(Dense(1, kernel_initializer=initializer))

model.compile(loss='mean_squared_error', optimizer='adam')

# Train the model
model.fit(X, Y, epochs=500, verbose=0)

# Predict the output
Y_pred = model.predict(X)

# Plot the original function and the model's predictions
plt.figure(figsize=(10,15))
plt.subplot(4, 2, 1)
plt.plot(X, Y, label='Original Function')
plt.plot(X, Y_pred, label='Neural Network Output')
plt.legend()

# Show the activations of each neuron in the hidden layer for the input range
hidden_layer_weights = model.layers[0].get_weights()[0]
hidden_layer_biases = model.layers[0].get_weights()[1]

for i in range(hidden_layer_weights.shape[1]):
    plt.subplot(4, 2, i+2)
    neuron_output = np.maximum(0.01 * X.dot(hidden_layer_weights[:,i]) + hidden_layer_biases[i], X.dot(hidden_layer_weights[:,i]) + hidden_layer_biases[i])
    plt.plot(X, neuron_output, label=f'Neuron {i+1} Output')
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense

# Create a simple dataset
x = np.linspace(-1, 1, 100)
y = x**3

# Create a neural network with 5 neurons in the hidden layer
model = Sequential([
    Dense(5, input_dim=1, activation='sigmoid', kernel_initializer='he_normal'),
    Dense(1, activation='linear', kernel_initializer='he_normal')
])

# Compile the model
model.compile(loss='mean_squared_error', optimizer='adam')

# Train the model
model.fit(x, y, epochs=500, verbose=0)

# Get the weights of the hidden layer
weights, biases = model.layers[0].get_weights()

# Create the output of each neuron in the hidden layer
neuron_outputs = 1 / (1 + np.exp(-x.reshape(-1,1) @ weights - biases))

# Create a subplot for each neuron output
fig, axs = plt.subplots(3, 2, figsize=(10, 15))

for i in range(5):
    axs[i // 2, i % 2].plot(x, neuron_outputs[:, i])
    axs[i // 2, i % 2].set_title(f'Neuron {i+1}')

# Remove the extra subplot (if any)
if axs.size > 5:
    fig.delaxes(axs.flatten()[5])

# Show the final output of the model
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='True function')
plt.plot(x, model.predict(x), label='Neural network output')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense

# Create a simple dataset
x = np.linspace(-1, 1, 100)
y = x**3

# Create a neural network with 5 neurons in the hidden layer
model = Sequential([
    Dense(5, input_dim=1, activation='relu', kernel_initializer='he_normal'),
    Dense(1, activation='linear', kernel_initializer='he_normal')
])

# Compile the model
model.compile(loss='mean_squared_error', optimizer='adam')

# Train the model
model.fit(x, y, epochs=500, verbose=0)

# Get the weights of the hidden layer
weights, biases = model.layers[0].get_weights()

# Create the output of each neuron in the hidden layer
neuron_outputs = np.maximum(0, x.reshape(-1,1) @ weights + biases)

# Create a subplot for each neuron output
fig, axs = plt.subplots(3, 2, figsize=(10, 15))

for i in range(5):
    axs[i // 2, i % 2].plot(x, neuron_outputs[:, i])
    axs[i // 2, i % 2].set_title(f'Neuron {i+1}')

# Remove the extra subplot (if any)
if axs.size > 5:
    fig.delaxes(axs.flatten()[5])

# Show the final output of the model
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='True function')
plt.plot(x, model.predict(x), label='Neural network output')
plt.legend()
plt.show()
```
