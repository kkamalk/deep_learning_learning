### The important point showcased in this notebook is the effect of learning rate on the convergence of the learning algorithm.

## Learing rate = 0.1

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a weight array
weights = np.linspace(-20, 20, 400)

# Convex cost function: cost = (weight - optimal_weight)^2
# Set optimal weight such that cost is minimized (for example, optimal_weight = 3)
costs = (weights - 3)**2

# Plot the cost function
plt.figure(figsize=(8,6))
plt.plot(weights, costs, label='Cost function')

# Simulate batch gradient descent
# Let's start with an initial weight of -10
current_weight = -10
# Learning rate for the gradient descent
learning_rate = 0.1  # Increase the learning rate to make it 'bounce around'
# Store weights for plotting
weight_history = [current_weight]

for _ in range(20):  # Do 20 iterations
    # Compute gradient at current weight: gradient = 2*(weight - optimal_weight)
    gradient = 2 * (current_weight - 3)
    # Update weight
    current_weight -= learning_rate * gradient
    # Store weight
    weight_history.append(current_weight)
    current_cost = (current_weight - 3)**2
    print("weight:  " ,current_weight, "   gradient:   ", gradient, "     Cost:    ", current_cost)

# Plot weight updates
plt.plot(weight_history, [(w - 3)**2 for w in weight_history], 'ro-', label='Batch Gradient Descent')

plt.xlabel('Weight')
plt.ylabel('Cost')
plt.legend()
plt.show()
```
